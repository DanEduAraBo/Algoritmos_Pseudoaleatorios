```python
import matplotlib.pyplot as plt
import random
import math
```

```python
def obtener_a(m):
    t = random.randrange(1, math.floor(m/9))
    r = random.random()
    if r < 0.5:
        a = 8*t - 3
    else:
        a = 8*t + 3
    return a
```

```python
def obtener_x0(m):
    x0 = m
    while x0%2 == 0 or x0%m == 0:
        x0 = random.randrange(1, m)
    return (x0)
```

```python
def serie_random_mcm(cantidad_de_numeros, a, x0, m):
    X = x0
    serie = [X]
    for i in range(cantidad_de_numeros):
        Xn = (a*X) % m
        serie.append(Xn)
        X = Xn
    return serie
```

```python
d = 64
m = 2*d
a = obtener_a(m)
x0 = obtener_x0(m)
cantidad_de_numeros = math.ceil(m/4)+3
```

```python
numeros_aleatorios = serie_random_mcm(cantidad_de_numeros, a, x0, m)
plt.plot(list(range(len(numeros_aleatorios))), numeros_aleatorios)
plt.show()
```
