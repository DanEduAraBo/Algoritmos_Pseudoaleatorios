# Números pseudo-aleatorios

```python
import matplotlib.pyplot as plt
import random
```

# Método Congruencial Lineal

```python
def obtener_a(m):
    a = 3
    while a%3 == 0 or a%5 == 0 or a%2==0:
        a = random.randrange(1,m)
    return a
```

```python
def obtener_c(m):
    c = 8
    while c%8 != 5:
        c = random.randrange(1,m)
    return c
```

```python
def serie_random_mcl(cantidad_de_numeros, a, c, m, semilla=None):
    if semilla is None:
        semilla = random.randrange(1,m+1)
    X = semilla
    serie = [X]
    for i in range(cantidad_de_numeros):
        Xn = (a*X + c) % m
        serie.append(Xn)
        X = Xn
    return serie
```

```python
p = 2
d = 64
cantidad_de_numeros = 130
m = 127
a = obtener_a(m)
c = obtener_c(m)
```

```python
numeros_aleatorios = serie_random_mcl(cantidad_de_numeros, a, c, m)
plt.plot(list(range(len(numeros_aleatorios))), numeros_aleatorios)
plt.show()
```

## Método Congruencial Lineal Uniforme

```python
numeros_aleatorios_uniformes = [x/m for x in numeros_aleatorios]
plt.plot(list(range(len(numeros_aleatorios_uniformes))), numeros_aleatorios_uniformes)
plt.show()
```
